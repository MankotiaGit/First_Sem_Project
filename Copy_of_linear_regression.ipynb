```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/placement (1).csv')
df
```

```python
#data analysis
df.info()#to find info
```

```python
#find the null value
df.isnull().sum()
```

```python
#to know how data look mathematically
df.describe()
```

```python
#to find no of columns and rows
df.shape
```

```python
#plot the data
plt.scatter(df['cgpa'],df['package'])
plt.xlabel('cgpa')
plt.ylabel('package')
plt.show()
```

```python
#find the dependent and independent feature
x = df.iloc[:,:0]#independent feature
y = df.iloc[:,1]#dependent feature
```

```python
x
```

```python
#split the data int train and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)#random state fix the suffling process and give fix value
x_train
```

```python
#import the model and training
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x_train.values.reshape(-1, 1) ,y_train)
```

```python
m = model.coef_
c = model.intercept_
print(m,c)
```

```python
cgpa =8.5
package = m * cgpa + c
print(package)
```

```python
#prediction the model
y_pred = model.predict(x_test.values.reshape(-1, 1))
y_pred
```
